```python
import pandas as pd
import numpy as np

csv_path = "./final_data.csv"
```

```python
df = pd.read_csv(csv_path, header=None)
target = df.columns[-1]
```

```python
df.describe()
```

```python
# Find correlations between all columns and the last column
correlations_matrix = df.corr()
correlations = correlations_matrix[target]

# Print the correlations
print("Correlation of each variable with the last column:")
print(correlations)
```

```python
len(np.where(correlations.isna())[0]) #Pawns + Kings
```

```python
correlations_not_nan = np.nan_to_num(correlations, nan=0)

# Get the indices of the sorted array
sorted_indices = np.argsort(correlations_not_nan)[::-1]
# Use the indices to get the sorted values
sorted_values = correlations_not_nan[sorted_indices]

# Print the results
print("Per blanques")
print("Sorted indices:", sorted_indices)
print("Sorted values:", sorted_values)
```
